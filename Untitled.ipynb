```python
for x in range(baris):
    for y in range(kolom):
        max1=0
        max2=0
        for z in range(3):
            if(max1==0):
                max1=foto2[x,y,z]
                imax1=z
            elif(max1<foto2[x,y,z]):
                max1=foto2[x,y,z]
                imax1=z
            elif(max2<foto2[x,y,z]):
                max2=foto2[x,y,z]
                imax2=z
        if((foto2[x,y,imax1]-foto2[x,y,imax2])>10):
            foto2[x,y,imax1]=foto2[x,y,imax1]-foto2[x,y,imax2]/2
            for z in range(3):
                if(z!=imax1):
                    foto2[x,y,z]=0

plt.imshow(foto2)
```

Anda menggunakan dua loop bersarang (for x in range(baris) dan for y in range(kolom)) untuk mengakses setiap piksel dalam gambar.

Di dalam loop tersebut, Anda mencari dua nilai terbesar (max1 dan max2) dari nilai warna dalam saluran warna (R, G, B) di setiap piksel.

Kemudian, Anda membandingkan kedua nilai tersebut. Jika perbedaan antara nilai terbesar pertama dan kedua lebih besar dari 10, Anda mengurangi setengah dari nilai terbesar kedua dari nilai terbesar pertama.
